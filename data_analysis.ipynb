```python
import pandas as pd
df = pd.read_csv('Used_Bikes.csv')
df
```

```python
text = "File name : Used_Bikes.csv \n the dataset contain details of used bikes and its curresponding price\n"
with open('data_dictionary.txt','a')as fp:
    fp.write(text)
    fp.write(f"size of data \n rows : {df.shape[0]} \n columns : {df.shape[1]} \n total size : {df.size}\n\n")
    fp.write("columns in dataset and its datatype:\n")
    dtype =  ['string' if str(i) == 'object' else str(i) for i in df.dtypes]
    for i,j in zip(df.columns,dtype):
        fp.write(f"{i}({j}) : {df.shape[0]-df['price'].notnull().sum()} null values found\n")
```

```python
df['bike_name'] =df.apply(lambda row: row['bike_name'].replace(row['brand'],'').strip(),axis=1)
```

```python
for i in ['bike_name','city','owner','brand']:
    df[i] = df[i].str.lower()
```

```python
df['owner'].replace({'first owner':1,'second owner':2,'third owner':3,'fourth owner or more':4},inplace=True)
```

```python
df
```

```python
import category_encoder as ce
binary_encoder = ce.BinaryEncoder(cols=['bike_name','city','brand'])
bike_encoded = binary_encoder.fit_transform(df.drop(columns=['price']))
bike_encoded
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(bike_encoded,df['price'],test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
}
grid_search = GridSearchCV(estimator=RandomForestRegressor(),
                           param_grid=param_grid,
                           cv=5)  # Verbosity level
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
model = RandomForestRegressor(n_estimators=300).fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
df
```

```python
test_data = pd.DataFrame({'bike_name':['classic 350cc'],'city':['delhi'],'kms_driven':[110],'owner':[1],'age':[1.0],'power':[450.0],'brand':['royal enfield']})
test_data_encoded = binary_encoder.transform(test_data)
model.predict(test_data_encoded)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
cv_score_train = cross_val_score(model,x_train,y_train)
cv_score_test = cross_val_score(model,x_test,y_test)
r2score = r2_score(df['price'],model.predict(bike_encoded))
```

```python
model_eval_result = {'cv_score_for_training':cv_score_train,'cv_score_for_test':cv_score_test,'r2score':r2score}
model_eval_result
```
